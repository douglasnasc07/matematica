### Circunferência:

- Equação reduzida:

\begin{equation}
(x - x_0)^2 + (y - y_0)^2 = R^2
\end{equation}

\begin{equation}
x = \pm \sqrt{R^2 - (y - y_0)^2} + x_0
\end{equation}

\begin{equation}
y = \pm \sqrt{R^2 - (x - x_0)^2} + y_0
\end{equation}

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-4.0, 6.0, 100)

def coord_y1(R,x0,y0):
    return -1.0*np.sqrt(R**2 - (x - x0)**2) + y0

def coord_y2(R,x0,y0):
    return 1.0*np.sqrt(R**2 - (x - x0)**2) + y0

```

```python
y1 = coord_y1(5.0, 1.0, 1.0) #Raio 5.0 e centro C(1.0 , 1.0)
y2 = coord_y2(5.0, 1.0, 1.0)
```

```python
plt.figure(figsize=(5, 5))
plt.plot(x,y1)
plt.plot(x,y2)
plt.grid(True)
```

- Equação paramétrica:

\begin{equation}
\vec{f}(t) = (x_0 + R cos (t), y_0 + R sen (t))
\end{equation}

```python
t = np.linspace(-1, 1)

def coord_x(x0,R):
    return x0 + (R*np.cos(np.pi*t))

def coord_y(y0,R):
    return y0 + (R*np.sin(np.pi*t))



```

```python
x = coord_x(0.0, 1.0) #Raio 1.0 e centro (0,0)
y = coord_y(0.0, 1.0)

plt.figure(figsize=(5, 5))
plt.plot(x,y)
plt.grid(True)
```

```python
x = coord_x(-2.0, 1.0)  #Raio 1.0 e centro (-2.0, 4.0)
y = coord_y(4.0, 1.0)

plt.figure(figsize=(5, 5))
plt.plot(x,y)
plt.grid(True)
```

### Elipse:

- Equação reduzida:

\begin{equation}
\frac{(x - x_0)^2}{a^2} + \frac{(y - y_0)^2}{b^2} = 1
\end{equation}

\begin{equation}
x = \pm \sqrt{a^2 - \frac{a^2}{b^2}(y - y_0)^2} + x_0
\end{equation}

\begin{equation}
y = \pm \sqrt{b^2 - \frac{b^2}{a^2}(x - x_0)^2} + y_0
\end{equation}

```python
x = np.linspace(-4.999999, 4.999999, 100)

def coord_y1(a,b,x0,y0):
    return -1.0*np.sqrt(b**2 -((b/a)**2)* (x - x0)**2) + y0

def coord_y2(a,b,x0,y0):
    return 1.0*np.sqrt(b**2 -((b/a)**2)* (x - x0)**2) + y0
```

```python
y1 = coord_y1(5.0, 1.0, 0.0, 0.0) #a = 5.0, b = 1.0, x0 = 0.0, y0 = 0.0
y2 = coord_y2(5.0, 1.0, 0.0, 0.0)
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.grid(True)
```

- Equação paramétrica:

\begin{equation}
\vec{f}(t) = (x_0 + a cos (t), y_0 + b sen (t))
\end{equation}

```python
t = np.linspace(-1, 1)

def coord_x(x0,a):
    return x0 + (a*np.cos(np.pi*t))

def coord_y(y0,b):
    return y0 + (b*np.sin(np.pi*t))
```

```python
x = coord_x(0.0, 5.0) #a = 5.0, b = 1.0, x0 = 0.0, y0 = 0.0)
y = coord_y(0.0, 1.0)

plt.plot(x,y)
plt.grid(True)
```

### Parábola:

- Equação reduzida:

\begin{equation}
ax^2 + dy + bx + c =0
\end{equation}

```python
x = np.linspace(-6.0, 6.0, 100)

def coord_y(a,b,c,d):
    return (a*x**2 + b*x + c)/d

y = coord_y(-1.0 , 1.0 , 1.0 , 1.0)

plt.figure(figsize=(5, 5))
plt.plot(x,y)
plt.grid(True)

```

- Equação paramétrica:

\begin{equation}
x = t
\end{equation}

\begin{equation}
y = at^2 + bt + c
\end{equation}

```python
t = np.linspace(-6.0, 6.0, 100)

def coord_y(a,b,c,d):
    return (a*t**2 + b*t + c)/d

x = t
y = coord_y(-1.0 , 1.0 , 1.0 , 1.0)

plt.figure(figsize=(5, 5))
plt.plot(x,y)
plt.grid(True)
```

### Hipérbole:

- Equação reduzida:

\begin{equation}
\frac{(x - x_0)^2}{a^2} - \frac{(y - y_0)^2}{b^2} = 1
\end{equation}

\begin{equation}
x = \pm \sqrt{a^2 + \frac{a^2}{b^2}(y - y_0)^2} + x_0
\end{equation}

\begin{equation}
y = \pm \sqrt{b^2 + \frac{b^2}{a^2}(x - x_0)^2} + y_0
\end{equation}

```python
x = np.linspace(-10.0, 10.0, 100)

def coord_y1(a,b,x0,y0):
    return -1.0*np.sqrt(b**2 +((b/a)**2)* (x - x0)**2) + y0

def coord_y2(a,b,x0,y0):
    return 1.0*np.sqrt(b**2 +((b/a)**2)* (x - x0)**2) + y0
```

```python
y1 = coord_y1(5.0, 1.0, 0.0, 0.0) #a = 5.0, b = 1.0, x0 = 0.0, y0 = 0.0
y2 = coord_y2(5.0, 1.0, 0.0, 0.0)
```

```python
plt.xlim(-10, 10)  
plt.ylim(-2.5, 2.5) 
plt.plot(x,y1)
plt.plot(x,y2)
plt.grid(True)
```

- Equação paramétrica:

\begin{equation}
x = x_0 +b senh(t)
\end{equation}

\begin{equation}
y = y_0 \pm a cosh(t)
\end{equation}

\begin{equation}
\vec{f}(t) = (x_0 +b senh(t),  y_0 \pm a cosh(t))
\end{equation}

```python
t = np.linspace(-10.0, 10.0, 100)

def coord_y1(x0,a):
    return x0 - a*np.cosh(t)

def coord_y2(x0,a):
    return x0 + a*np.cosh(t)

def coord_x(y0,b):
    return y0 + b*np.sinh(t)
```

```python
y1 = coord_y1(0.0, 1.0)
y2 = coord_y2(0.0, 1.0)
x = coord_x(0.0, 5.0)
```

```python
plt.xlim(-10, 10)  
plt.ylim(-2.5, 2.5) 
plt.plot(x,y1)
plt.plot(x,y2)



plt.grid(True)
```
